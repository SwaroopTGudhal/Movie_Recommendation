```python
import pandas as pd
import lxml
from urllib.request import urlopen as uReq
from bs4 import BeautifulSoup as bs
import requests
import html5lib
import numpy as np
```

```python
def get_urls(movie_type="American", Select_start_year=2019, End_year=2020):
    urls = []
    for year in range(Select_start_year, End_year+1):
        url = "https://en.wikipedia.org/wiki/List_of_{}_films_of_{}".format(movie_type,year)
        urls.append(url)
    return urls
```

```python
url_list = get_urls(movie_type="American", Select_start_year=2010, End_year=2020)
url_list
```

```python
def get_table(url):
    wikiurl=url
    table_class="wikitable sortable"
    response=requests.get(wikiurl)
    soup = bs(response.text, 'html.parser')
    indiatable=soup.find_all('table',{'class':table_class})
    df=pd.read_html(str(indiatable))
    df1 = pd.DataFrame(df[0])
    df2 = pd.DataFrame(df[1])
    df3 = pd.DataFrame(df[2])
    df4 = pd.DataFrame(df[3])
    df_url = df1.append(df2.append(df3.append(df4,ignore_index=True),ignore_index=True),ignore_index=True)
    return df_url
```

```python
def get_director(x):
    if " (director)" in x:
        return x.split(" (director)")[0]
    elif " (directors)" in x:
        return x.split(" (directors)")[0]
    elif " (directors/screenplay)" in x:
        return x.split(" (directors/screenplay)")[0]
    else:
        return x.split(" (director/screenplay)")[0]
    
    
def get_actor1(x):
    return ((x.split("screenplay); ")[-1]).split(", ")[0])



def get_actor2(x):
    if len((x.split("screenplay); ")[-1]).split(", ")) < 2:
        return np.NaN
    else:
        return ((x.split("screenplay); ")[-1]).split(", ")[1])
    

def get_actor3(x):
    if len((x.split("screenplay); ")[-1]).split(", ")) < 3:
        return np.NaN
    else:
        return ((x.split("screenplay); ")[-1]).split(", ")[2])
```

```python
data = pd.DataFrame()
for url in url_list:
    x = get_table(url)
    data = data.append(x)
data
```
